# So Neural Networks are universal function approximators?

__Goals__ 

The [Universal approximation theorem](https://en.wikipedia.org/wiki/Universal_approximation_theorem) states that a feed-forward network with a single hidden layer containing a finite number of neurons can approximate continuous functions on compact subsets of R^n, under mild assumptions on the activation function. 

In this notebook we seek to verify this result, by training neural networks to approximate a number of common mathematical functions using synthetic examples.

In particular, we are interested in:
* How deep/wide does the NN need to be in practice to approximate such functions?
* What difference does using different activation functions bring?
* What kind of weights does the NN end up getting (esp. for a simple function like f(x) = x)? How does this change if we apply reguarization?

__Function classes to be approximated__
1. Elementary math functions ([0, 1] -> R)
    * c + x (for these it'd be interesting to see what kinds of weights the nn ends up getting, especially when we use multiple layers)
    * c - x
    * c * x
    * c / x
    * x ^ c
    * exp(x)
    * log(x)
    * x^(1/n)
    * sin(x) (not elementary, but ok..)
    * cos(x)
2. Boolean logic ((boolean, boolean) -> boolean)
    * x AND y
    * x OR y
    * not x
3. Sequence patterns (for RNN)? (boolean sequence -> boolean sequence)

# Elementary math functions

## Code to generate data, train nn, and plot results

* We investigate functions from [1, 2] -> R.
    * Not using [0, 1] -> R to avoid the Inf behavior of 1/x near 0
* We use mean absolute percentage error (MAPE) as the error metric, to account for the varying range of different functions.
    * Let's call mape < 0.1 "a good fit".
* Loss function is mean squared error.

```python
# Class to hold configs
class Config(object):
    def __init__(self,  
                 fun, # function to be approximated
                 fun_name, # function name (for plotting purporse)
                 sample_size = 10000, # number of synthetic examples generated
                 layer1_nodes=10, # number of nodes in hidden layer 1
                 layer2_nodes=None, # number of nodes in hidden layer 2; if None no hidden layer 2 
                 layer3_nodes=None, # number of nodes in hidden layer 3; if None no hidden layer 3
                 activation="relu", # activation function (for all layers)
                 epochs=50,
                 batch_size=16,
                 optimizer="adagrad",
                 verbose=1 # 0 for no logging to stdout, 1 for progress bar logging, 2 for one log line per epoch
                ):
        self.fun = fun # 
        self.fun_name = fun_name
        self.sample_size = sample_size
        self.layer1_nodes = layer1_nodes
        self.layer2_nodes = layer2_nodes
        self.layer3_nodes = layer3_nodes
        self.activation = activation
        self.epochs = epochs
        self.batch_size = batch_size
        self.optimizer = optimizer
        self.verbose = verbose
```

```python
# Function to generate synethic data, train a nn, and plot result
def approximate(config, data=datasets):
    np.random.seed(123)
    # Generate synthetic data
    examples = np.random.uniform(1.0, 2.0, size=config.sample_size)
    targets = config.fun(examples)
    
    # Define a NN model
    def create_model():
        nn = Sequential()
        nn.add(Dense(config.layer1_nodes, input_dim=1, activation=config.activation)) # Layer 1
        if config.layer2_nodes is not None:   
            nn.add(Dense(config.layer2_nodes, activation="relu"))   # Optional layer 2
            if config.layer3_nodes is not None:
                nn.add(Dense(config.layer3_nodes, activation="relu")) # Optional layer 3
        nn.add(Dense(1)) # Output layer
        nn.compile(optimizer=config.optimizer,loss='mean_squared_error', metrics=["mean_absolute_percentage_error"])
        return nn
    
    # Train NN model on synthetic data
    def train_model():
        model = create_model()
        history = model.fit(examples, targets, epochs=config.epochs,
                            batch_size=config.batch_size, verbose=config.verbose)
        return model, history
    
    # Plot output
    def plot_result(preds, history):
        sorted_examples = examples[examples.argsort()]
        sorted_targets = targets[examples.argsort()]
        sorted_preds = preds[examples.argsort()]
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
        ax1.plot(sorted_examples, sorted_targets, color="royalblue", label="true")
        ax1.plot(sorted_examples, sorted_preds, color="green", label="fitted")
        ax1.set_title("True and fitted function values", fontsize=14)
        ax1.set_xlabel("x")
        ax1.set_ylabel("f(x)")
        errors = history.history['mean_absolute_percentage_error'][3:]
        epochs = history.epoch[3:]
        ax2.plot(epochs, errors)
        ax2.set_title("Mean absolute percentage error after each epoch (4th onwards)", fontsize=14)
        ax2.set_xlabel("epoch")
        ax2.set_ylabel("mape")
        ax2.set_xlim()
        final_mape = round(history.history['mean_absolute_percentage_error'][-1], 4)
        fig.suptitle("F(x)={}, layer1={}, layer2={}, layer3={}, activation={}, optimizer={}. Final MAPE={}".format(config.fun_name, 
             config.layer1_nodes, config.layer2_nodes, config.layer3_nodes, 
            config.activation, config.optimizer, final_mape),
                     fontsize=16)
        ax1.legend()
        fig.show()
        return fig

    model, history = train_model()
    preds = model.predict(examples)
    fig = plot_result(preds, history)
    return model, fig
```

## Division

### Dense(10)

```python
config = Config(fun=lambda x:1/x, 
                fun_name="1/x", 
                layer1_nodes=10,
                layer2_nodes=None,
                layer3_nodes=None,
                activation="relu",
                verbose=0)
model, fig = approximate(config)
```

The NN with 1 Dense(10) layer doesn't approximate 1/x very well. MAPE stagnated after around 2.9. It basically fitted a straight line.

What about making the NN wider (increase hidden layer size)?

### Dense(1000)

```python
config = Config(fun=lambda x:1/x, 
                fun_name="1/x", 
                layer1_nodes=1000,  # 100 nodes instead
                layer2_nodes=None,
                layer3_nodes=None,
                activation="relu",
                verbose=0)
model, fig = approximate(config)
```

Using a 1000-nodes layer didn't make the NN better. Looking at the chart on the left, it is still basically just fitting a straight line.

(The jumpyness is somewhat misleading - it is due to the small y axis range.)

Train longer?

### Dense(1000), 200 epochs

```python
config = Config(fun=lambda x:1/x, 
                fun_name="1/x", 
                layer1_nodes=1000,  # 1000 nodes instead
                layer2_nodes=None,
                layer3_nodes=None,
                activation="relu",
                epochs=200,        # patience
                verbose=0)
model, fig = approximate(config)
```

Not much change.

### Dense(1000), adam optimizer

```python
config = Config(fun=lambda x:1/x, 
                fun_name="1/x", 
                layer1_nodes=1000, 
                layer2_nodes=None,
                layer3_nodes=None,
                activation="relu",
                optimizer="adam",   # adam optimizer instead of adagrad
                epochs=200,
                verbose=0)
model, fig = approximate(config)
```

Using the Adam optimizer actually made a difference! MAPE is slowly creeping towards 0.1. The fitted curve looks a lot more reasonable now.

What about making the NN deeper instead?
